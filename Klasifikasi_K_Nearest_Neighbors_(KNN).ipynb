```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. LOAD DATASET
df = pd.read_csv('smartprix_smartphones_april_2026.csv')

# 2. SELEKSI FITUR & TARGET
X = df[['price', 'spec_score']]
y = df['has_5G'].astype(int) # Mengubah True/False menjadi 1/0

# 3. NORMALISASI (SCALING)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. SPLIT DATA
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 5. MEMBUAT MODEL KNN
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)

# 6. PREDIKSI
y_pred = model_knn.predict(X_test)

# 7. EVALUASI HASIL
print("======= HASIL PRAKTIKUM KNN =======")
print(f"Akurasi Model: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nLaporan Klasifikasi:")
print(classification_report(y_test, y_pred, target_names=['Bukan 5G', 'Sudah 5G']))

# 8. PREDIKSI DATA BARU (Manual)
harga_baru = 25000
skor_baru = 75
data_baru = [[harga_baru, skor_baru]]

data_baru_scaled = scaler.transform(data_baru)
hasil_prediksi = model_knn.predict(data_baru_scaled)

label = "Mendukung 5G" if hasil_prediksi[0] == 1 else "Tidak Mendukung 5G"
print(f"\nPrediksi untuk HP Harga {harga_baru} & Skor {skor_baru}: {label}")
```

**1. Penjelasan Kode (Langkah demi Langkah)**

**Persiapan Data & Library**
  * **Import Library**: Menggunakan pandas untuk olah data, sklearn untuk mesin pembelajarannya, dan StandardScaler untuk menyamakan skala angka.
  * **Load Dataset**: Membaca file CSV. Pastikan file tersebut berada di folder yang sama dengan skrip kamu.
  * **Fitur (X) & Target (y)**:
    * X adalah variabel independen (harga dan skor spek) yang digunakan untuk menebak.
    *  y adalah target (apakah HP itu 5G atau tidak). Kita mengubahnya menjadi angka (0 dan 1) agar komputer bisa menghitungnya secara matematis.
  
**Normalisasi (Scaling)**
  * **Pentingnya Scaling**: KNN sangat sensitif terhadap jarak. Karena price (misal: 25.000) punya angka jauh lebih besar dibanding spec_score (misal: 75), tanpa scaling, model akan menganggap harga jauh lebih penting daripada skor. StandardScaler membuat keduanya memiliki rentang nilai yang setara.

**Split Data & Pelatihan**
  * **Train/Test Split**: Membagi data menjadi 80% untuk belajar dan 20% untuk ujian. Ini penting agar kita tahu apakah model benar-benar pintar atau cuma "menghafal" data lama.
  * **K=5**: Model akan melihat 5 tetangga terdekat sebelum mengambil keputusan.

**Evaluasi & Prediksi**
  * **Classification Report**: Memberikan detail presisi (ketepatan) dan recall (seberapa banyak yang berhasil ditangkap).
  * **Prediksi Data Baru**: Kamu melakukan Scaling ulang pada data baru sebelum dimasukkan ke model. Ini langkah krusial; jika data latihan di-scale, data baru pun wajib di-scale dengan parameter yang sama.

**2. Skenario Pengujian (Test Scenarios)**

Untuk memastikan model KNN kamu berjalan optimal dan valid, kamu bisa melakukan pengujian dengan skenario berikut:

**A. Pengujian Nilai K (K-Fold Cross Validation)**

Jangan terpaku pada $k=5$.

  * **Skenario**: Coba ubah nilai n_neighbors menjadi angka ganjil lain (misal: 1, 3, 7, 9).
  * **Ekspektasi**: Jika K terlalu kecil (1), model rentan terhadap noise (overfitting). Jika K terlalu besar, model menjadi terlalu umum (underfitting). Cari nilai K yang memberikan akurasi stabil pada data uji.
  
**B. Pengujian Outlier (Data Ekstrim)**

  * **Skenario**: Masukkan data HP dengan harga sangat murah (misal: 1.000) tapi spek sangat tinggi (misal: 95), atau sebaliknya.
  * **Ekspektasi**: Lihat apakah KNN terkecoh atau tetap konsisten mengikuti pola data yang sudah ada di sekitar titik tersebut.
  
**C. Pengujian Rasio Split Data**

  * **Skenario**: Ubah test_size dari 0.2 menjadi 0.3 atau 0.4.
  * **Ekspektasi**: Mengamati apakah penurunan jumlah data latihan (training data) drastis mempengaruhi akurasi. Jika akurasi turun drastis, artinya model butuh lebih banyak data untuk belajar.

**D. Pengujian Matriks Kebingungan (Confusion Matrix)**

Gunakan fungsi yang sudah di-import: print(confusion_matrix(y_test, y_pred)).

    * **Skenario**: Analisis berapa kali model salah menebak HP 5G sebagai HP Biasa (False Negative).
    * **Ekspektasi**: Dalam dunia nyata, salah menebak HP mahal tidak punya 5G mungkin lebih "fatal" bagi pembeli daripada salah menebak HP murah punya 5G.
